```python
import os
import json
class students:
  def __init__(self,id,name,age,grade):
    self.id=id
    self.name=name
    self.age=age
    self.grade=grade
  def __str__(self):
    return f"ID:{self.id}, Name:{self.name}, Age:{self.age}, Grade:{self.grade}"
class studentmanager:
  def __init__(self):
    self.students = []
    self.filename ="student.jason"
    self.loadfromfile()
  def add(self):
    id=int(input("Enter ID:"))
    name=str(input("Enter Name:"))
    age=int(input("Enter Age:"))
    grade=str(input("Enter Grade:"))
    student = students(id, name, age, grade)
    self.students.append(student)
    self.save_to_file()
    print("Student Added Successfully")
  def view(self):
    for student in self.students:
      print(student)
      return
    print("No record found")
  def search(self):
    sid = int(input("Enter ID to search:"))
    for student in self.students:
      if student.id == sid:
        print(student)
        return
      print("No record found")
  def delete(self):
    sid = int(input("Enter ID to search:"))
    for student in self.students:
      if student.id == sid:
        self.students.remove(student)
        print("Student Data Deleted successfully")
        return
    print("No record found")
  def save_to_file(self):
    with open(self.filename,"w") as f:
      data=[vars(student) for student in self.students]
      json.dump(data,f)
    print("Saved successfully")
  def loadfromfile(self):
    if os.path.exists(self.filename):
      with open(self.filename,"r") as f:
        try:
         data = json.load(f)
         self.students = (students(**item) for item in data) # Changed Student to students
        except json.JSONDecodeError:
         self.students = []
        else:
         self.students = []

  def main():
    manager = studentmanager()
    while True:
        print("1. Add student")
        print("2. View Student")
        print("3. Search Student")
        print("4. Delete ")
        print("5. Save & Exit")
        print("Enter your choice:")
        choice = int(input())
        if (choice == 1):
          manager.add()
        elif (choice == 2):
          manager.view()
        elif (choice == 3):
          manager.search()
        elif (choice == 4):
          manager.delete()
        elif (choice == 5):
          manager.save_to_file()
          print("üëã Exiting... Goodbye!")
          break
        else:
          print("‚ùó Invalid choice. Please try again.")
def main():
    # Your program logic here
    print("Hello from main!")

if __name__ == "__main__":
  s1=studentmanager()
  studentmanager.main()
```
